```python
import pandas as pd
```

```python
url = 'Data/City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month - City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
id_vars = df.columns[:8]
id_vars
```

```python
largest_cities = pd.melt(df[df['SizeRank'].isin([0,1,2,3])], id_vars=id_vars, var_name='Date', value_name='Value')
largest_cities
```

```python
import datetime
```

```python
largest_cities['Date'] = pd.to_datetime(largest_cities['Date'], format='%Y-%m-%d')
```

```python
largest_cities.dtypes
```

```python
largest_cities
```

```python
largest_cities = largest_cities.set_index('Date')
largest_cities
```

```python
largest_cities.index
```

```python
ts = largest_cities.groupby('RegionName').resample('M').sum()
```

```python
ts
```

```python
unstacked = ts.unstack(level=0)
unstacked
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def thousands_formatter(x,pos):
    return f'{x/1000:0.f}K'
```

```python
ax = unstacked['Value'].plot()
ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
```

```python
# Question 1
ts.loc[pd.IndexSlice[:,'12 2008'],:]
```

```python
# Question 2
chi_val = ts.loc['Chicago', 'Value']
hou_val = ts.loc['Houston', 'Value']
los_val = ts.loc['Los Angeles', 'Value']
new_val = ts.loc['New York', 'Value']
# make into DataFrame for .loc
comp_cols = pd.DataFrame({'Chicago':chi_val,
             'Houston':hou_val,
             'Los Angeles':los_val,
             'New York':new_val})
ts_compare = comp_cols.loc['11 2008':'12 2008']
ts_compare
```

```python
diff_compare = ts_compare.diff()
diff_compare
```
